```python
#import packages
import numpy as np
import scipy.io as io
```

```python
#load data
data= io.loadmat('fluxes.mat')
##display(data)
```

```python
#create constants
rhoa = np.float64(data['rhoa'])
rhoo = np.float64(data['rhoo'])
cs = np.float64(data['cs'])
cl = np.float64(data['cl'])
cpa = np.float64(data['cpa'])
cpo = np.float64(data['cpo'])
lE = np.float64(data['latentE'])
sB = np.float64(data['sigmaB'])

#create variables
sw = np.float64(data['sw'])
ws = np.float64(data['ws'])
lw = np.float64(data['lw'])
ta = np.float64(data['ta'])
ts = np.float64(data['ts'])
qa = np.float64(data['qa'])
```

## Problem 1

### A) code

using

Q=cp * rhoo * cpo * (h/dt)(T2-T1)

```python
#set variables for freezing temp (T1) and boiling (T2)
T1 = 0
T2 = 100
#find delta T
dT = T2-T1

#ocean depth var
h = 4000

#solar rad absorbed (Wm-2)
sr = 240

boiltime = (rhoo*cpo*h*dT)/sr
##print(boiltime, 'seconds')

#into years
s_per_yr = 60*60*24*365
boiltime_yrs = boiltime/s_per_yr
print(boiltime_yrs, 'years')
```

### C) code

```python
#set net heat
qnet = 0.8

#solve for dT/dt
warmrate = qnet/(rhoo*cpo*h)

##print(warmrate, "°C/s")

#convert to °C/decade
decade = 10
warmrate_d = warmrate*s_per_yr*decade
print(warmrate_d, "°C/decade")
```

### Complete Problem 1 Answers
A) 216 years

B) The oceans are not boiling as more is happening within the system beyond strictly shortwave absorption. There is significant longwave radiation from earth that releases heat energy as well as latent heat flux which removes energy from the ocean via evaporation

C) 0.015 °C/decade

#### b) 
The North Atlantic (60°N, 30°W) is SST is warmest in the northern hemisphere summer and coolest in the winter. ### C)

```python
fig, ax = plt.subplots(figsize=(7,4))

ax.plot(months, sw[:,0], marker='o', label='60°N 30°W')
ax.plot(months, sw[:,1], marker='s', label='0°N 180°W')
ax.legend()
ax.set_ylabel('Short Wave Radiation (Watts/$m^{2}$)')
ax.set_xlabel('Month')
ax.set_title('Annual Absorbed Short Wave Radiation')
```

### D)

Qs = Cp * rhoa * Cs * U (To - Ta)

```python
#calculate sensible heat flux (shf)
shf = -cpa*rhoa*cs*ws*(ts-ta)
##display(shf)

#plot shf
fig, ax = plt.subplots(figsize=(7,4))

#add zero reference line
plt.axhline(y=0, color='grey', linestyle='-')

ax.plot(months, shf[:,0], marker='o', label='60°N 30°W')
ax.plot(months, shf[:,1], marker='s', label='0°N 180°W')

ax.legend()
ax.set_ylabel('Sensible Heat Flux (Watts/$m^{2}$)')
ax.set_xlabel('Month')
ax.set_title('Annual Sensible Heat Flux')
```

This plot shows similarities with the in class plot. The North Atlantic has negative sensible heat flux with an average of -20 to -30 which is roughly the location of the data for the homework. The inc-class plot shows white (-10 to +10) in the equatorial pacific in which the calculated values match.

### E)

```python
#create humidity variable
qoex = (17.67 * ts)/(ts+243.5)
qo = 0.003802*np.exp(qoex)

dq = qo - qa
#calculate latent heat flux (shf)
lhf = -lE*rhoa*cl*ws*(dq)
##display(shf)

#plot lhf
fig, ax = plt.subplots(figsize=(7,4))

ax.plot(months, lhf[:,0], marker='o', label='60°N 30°W')
ax.plot(months, lhf[:,1], marker='s', label='0°N 180°W')

ax.legend()
ax.set_ylabel('Latent Heat Flux (Watts/$m^{2}$)')
ax.set_xlabel('Month')
ax.set_title('Annual Latent Heat Flux')
```

The equatorial pacific has more evaporation as the latent heat flux is more negative, corresponding to more heat energy leaving the ocean.

### F)

```python
#calculate upwelling longwave emissions for blackbody (ulwr)
ulwr = -sB*(ts+273.15)**4

#plot ulwr
fig, ax = plt.subplots(figsize=(7,4))

ax.plot(months, ulwr[:,0], marker='o', label='60°N 30°W')
ax.plot(months, ulwr[:,1], marker='s', label='0°N 180°W')

ax.legend()
ax.set_ylabel('Longwave Radiation (Watts/$m^{2}$)')
ax.set_xlabel('Month')
ax.set_title('Annual Blackbody Upwelling Longwave Emissions')

#find month of most emissions
lwnmin = min(ulwr[:,0])
lwnmin_month = np.where(ulwr[:,0]==lwnmin)
print(lwnmin_month)

lweqmin = min(ulwr[:,1])
lweqmin_month = np.where(ulwr[:,1]==lweqmin)
print(lweqmin_month)
```

There is the most blackbody radiation out of the ocean in the North Atlantic in August. There is the most blackbody radiation in the Equatorial Pacific in July

### G)

```python
#calculate downwelling longwave radiation (dlwr)
dlwr = lw-ulwr

#plot dlwr
fig, ax = plt.subplots(figsize=(7,4))

ax.plot(months, dlwr[:,0], marker='o', label='60°N 30°W')
ax.plot(months, dlwr[:,1], marker='s', label='0°N 180°W')

ax.legend()
ax.set_ylabel('Longwave Radiation (Watts/$m^{2}$)')
ax.set_xlabel('Month')
ax.set_title('Annual Downwelling Longwave Radiation')
```

### H)

```python
#calculate net surface heat flux
qnet = sw + lw + lhf + shf

#plot qnet
fig, ax = plt.subplots(figsize=(7,4))

ax.plot(months, qnet[:,0], marker='o', label='60°N 30°W')
ax.plot(months, qnet[:,1], marker='s', label='0°N 180°W')

ax.legend()
ax.set_ylabel('Net Heat Flux (Watts/$m^{2}$)')
ax.set_xlabel('Month')
ax.set_title('Annual Net Heat Flux')

#add zero reference line
plt.axhline(y=0, color='grey', linestyle='-')

nq_avg = np.mean(qnet[:,0])
eqq_avg = np.mean(qnet[:,1])

print(nq_avg)
print(eqq_avg)
```

a) For the N Atlantic, highest net surface heat flux in June and lowest in January
For the Eq Pacific, highest net surface heat flux in March and lowest in July

b) The values do not match and the temperature has max/mins about 3 months delayed from the net surface heat flux

c) Eq Pacific dominated by latent heat flux with greatest variation in seasonal patterns. N Atlantic dominated by shortwave radiation as has the most impactful seasonal cycle and magnitude.

d) The N Atlantic annual average net surface heat flux is -21.1 Watts/$m^{2}$ and the Eq Pacific is 83.7 Watts/$m^{2}$. This value does not have to equal zero as it is a single location. All of the global values summed should equal 0 for steady state, however Earth is not in steady state as it warms.
